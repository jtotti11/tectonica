```python
#Para estudiar los sismos y sus efectos debemos tener en cuenta que la peligrosidad de estos no solo esta definida por su magnitud sino 
#tambien por la profundidad a la que sucedieron.
#Primero queria registrar  como evoluciono la magnitud de los sismos durante el periodo 2004-2014 y posteriormente compararlo con la
#evolucion de la peligrosidad de los sismos durante el mismo periodo de tiempo. La peligrosidad de un evento sismico es la razon entre
#la profundidad del evento y su magnitud

peligrosidad=[]

x = np.zeros(len(cm))
w=list(range(len(cm)))
for i in range(len(x)):
    x[i]=w[i]

cm=np.zeros(3223)
for i in range(3223):
    cm[i]=cmagnitud[i]
    
cp=np.zeros(3223)
for i in range(3223):
    cp[i]=cprofundidad[i]
    
for i in range(len(cp)):
    c=cp[i]/cm[i]
    peligrosidad.append(c)
    
    
val= stats.linregress(x,cm)
val2 = stats.linregress(x,peligrosidad)
```

```python
figure(figsize=(20, 8))
scatter(x,cm, s=0.5, label="Sismos registrados")
plot(x,x*val[0]+val[1], lw=2, label="Ajuste lineal")
legend()
xlim([0,3500])
title("Evolucion magnitud de los sismos Cundinamarca(2004-2014)", fontsize=25)
xlabel("Longitud", fontsize=15)
ylabel("Latitud", fontsize=15)

print "Realice un regresion lineal para determinar como ha variado la magnitud sismica a lo largo de los 10 años."
print "Se observa que las magnitudes registradas han disminuido lo que parece un escenario conveniente con respecto"
print "a la segurida y prevencion de tragedias"
```

```python
figure(figsize=(20, 8))
scatter(x,peligrosidad,s=0.1, label="Sismos registrados")
plot(x,x*val2[0]+val2[1], lw=2, label="Ajuste lineal")
legend()
xlim([0,3500])
ylim([0,200])
title("Evolucion de la peligrosidad sismologica Cundinamarca(2004-2014)", fontsize=25)
xlabel("Longitud", fontsize=15)
ylabel("Latitud", fontsize=15)
print "Esta grafica cotradice de alguna manera la evolucion de magnitud ya que la peligrosidad de los eventos tiende a aumentar."
print "Esta figura nos muestra entonces una peligrosa relacion entre los eventos sismicos y su progreso en el tiempo"
```
